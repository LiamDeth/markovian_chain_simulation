## 1 - Paresse et échantillonnage

### 1.1 - Paresse

Soit $P$ une matrice de transition sur un espace d'états fini $E$ et $I$ la matrice identité sur $E$.
$\forall{N}\geq 1$, $P_N$ définie comme suit est une matrice de transition: 
\begin{equation} 
P_N = (1 - \frac{1}{N})I + \frac{1}{N}P = I + \frac{P-I}{N}
\end{equation} 
On dit que $(X^N_n)_{n \in \mathbb{N}}$ ayant $P_N$ comme matrice de transition pour $N\geq2$ est une chaîne paresseuse associée à $P$. Posons :
\begin{equation} 
Y_t^N = X^N_{\lfloor Nt \rfloor} \forall{t}\in\mathbb{R_+}
\end{equation} 
Alors, on a la convergence de $(Y^N)_{N\geq1}$ sur $E$:
\begin{equation} 
(Y^N)_{N\geq1} \rightarrow_{N \rightarrow \infty} (Y_t)_{t\geq0}
\end{equation} 
De plus, on peut exprimer pour $(Y_t)_{t\geq0}$:
* $A=P-I$ sont générateur
* $P_t = e^{tA}$ $\forall{t}\in\mathbb{R_+}$ son semi-groupe

### 1.2 - Echantillonnage

Réciproquement, si $(Y_t)_{t\in\mathbb{R_+}}$ est une chaîne de Markov à temps continu sur $E$ de générateur $A$, alors:
\begin{equation}
(Y_n^\delta)_{n\in\mathbb{N}} = (Y_{\delta n})_{n\in\mathbb{N}}
\end{equation}
est une chaîne de Markov à temps discret sur $E$ de matrice de transition $e^{\delta A}$ appelé chaîne échantillonnée de pas en temps $\delta$ par rapport à la chaîne de Markov à temps continue associée.

## 2 - Convergence d'une chaîne vers une diffusion

### 2.1 - Approximation diffusive

Soit $(P_N)_{N\geq1}$ une suite de noyaux de transition sur $\mathbb{R}$ et posons:
\begin{equation} 
b_N(x) = N \int_{|y-x|\leq1} {(y-x)P_N(x,dy)}
\end{equation} 
\begin{equation}
a_N(x) = N \int_{|y-x|\leq1} {(y-x)^2P_N(x,dy)}
\end{equation}
Supposons qu'il existe deux fonctions a et b à valeurs dans $]0,+\infty[$ et $\mathbb{R}$ respectivement, bornées et de classe $C^1$, et telles que $\forall{r}\geq0$ et $\forall{\epsilon}\geq0$,
\begin{equation}
\lim\limits_{N \to \infty} \sup\limits_{|x|\leq r} |b_N(x) - b(x)|= 0 ,   et    \lim\limits_{N \to \infty} \sup\limits_{|x|\leq r} |a_N(x) - a(x)|= 0
\end{equation}
et
\begin{equation}
\lim\limits_{N \to \infty} \sup\limits_{|x|\leq r} NP_N(x,[x-\epsilon,x+\epsilon]^c) = 0.
\end{equation}
Soit $(Y_n^N)_{n\in\mathbb{N}}$ la chaîne de Markov de noyau de transition $P_N$ et $(Z_t^N)_{t\in\mathbb{R_+}}$ le *processus* défini par $Z_t^N = Y_{\lfloor Nt \rfloor}^N$. Si $(Y_0^N)_{N\geq1}$ converge en loi vers $\nu$ alors, $\forall{T}\geq0$, on al aconvergence en loi:
\begin{equation}
(Z_t^N)_{t\in[0,T]} \rightarrow_{N \rightarrow \infty} (X_t)_{t\in[0,T]}
\end{equation}
Ou $(X_t)_{t\in[0,T]}$ est un processus de diffusion de loi initiale $\nu$ et de générateur A donné par:
\begin{equation}
Af(x) = \frac{1}{2}a(x)f^2(x) + b(x)f^1(x)
\end{equation}
Pour toute fonction f $C^2$ a support compact.

### 2.2 - Principe d'invariance (Donsker)

Soit $(Y_n)_{n\geq1}$ une suite de variables aléatoires i.i.d. telles que $\mathbb{E}[Y_1]=0$ et $\mathbb{V}ar[Y_1] = 1$, alors, si l'on note: 
\begin{equation}
X_t^N = \frac{1}{\sqrt{N}}(\sum_{k=1}^{\rfloor Nt \lfloor} Y_k + (N_t - \lfloor Nt \rfloor)Y_{\lfloor Nt \rfloor +1})
\end{equation}
La suite de processus $(X^N)_{N\geq1}$ converge en loi vers le mouvement brownien.

## 3 - Applications

### 3.1 - Marche Aléatoire

Soit $S_N$ la marche aléatoire définie sur $\mathbb{Z}$ $/ \forall n \in \mathbb{N}, S_N = \sum\limits_{i=1}^{n} \epsilon_i$

Ou $(\epsilon_i)_{i\in \mathbb{N}}$ est une suite de variables aléatoires i.i.d de loi $p_N\delta_1 + (1-p_N)\delta_{-1}$.


* Si $p_N = \frac{1}{2}$ alors le processus $(\frac{S_{\lfloor tN \rfloor}^{N}}{\sqrt{N}})_{t\geq0}$ converge en loi vers le mouvement brownien issue de 0.
* Si $p_N = p \neq \frac{1}{2}$ alors le processus $(\frac{S_{\lfloor tN \rfloor}^{N}}{N})_{t\geq0}$ converge en loi vers un processus déterministe.
* Si $p_N = \frac{1}{2} + \frac{\alpha}{2\sqrt{N}} $ alors le processus $(\frac{S_{\lfloor tN \rfloor}^{N}}{\sqrt{N}})_{t\geq0}$ converge vers le mouvement brownien à dérive $\alpha$ issue de 0.
 


```python
%matplotlib inline
!apt install ffmpeg

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib import animation

import numpy as np
import scipy.interpolate as sp

def anim_to_html(anim):
    plt.close(anim._fig)
    return anim.to_html5_video()

animation.Animation._repr_html_ = anim_to_html
```

```python
fig = plt.figure(figsize=(11, 11))
ax = fig.gca()

max_T = 5
N = 2
alpha = 2
PN = 0.5

T = np.linspace(0,max_T,500)

def XN(N,PN):
  return 2*np.random.binomial(1,PN,N)-1
  #return np.random.randn(N)
  
Positions = []
all_ = []

#ax.contour(X, Y, Z, alpha=.5)

def init():
    global Positions
    global N,PN,T,alpha
    global all_
    
    Positions = []
    all_ = []
    
    N = 2
    alpha = 0.
    PN = 0.5 + alpha/(2*np.sqrt(N))
    #PN = 0.25
    
    return all_
    
def update(frame):
    global Positions
    global N,PN,T
    global all_
    #[a.remove() for a in all_]
    #all_.clear()
    
    N *= 2
    PN = 0.5
    PN = 0.5 + alpha/(2*np.sqrt(N))
    pos = np.zeros(T.shape[0])
    X = XN(max_T*N,PN)
    for i in range(T.shape[0]):
      pos[i] = sum(X[:int(T[i]*N)]) / np.sqrt(N)
      
    all_ += ax.plot(T,pos)
    
    return all_

animation.FuncAnimation(fig, update, init_func=init, frames=15, interval=1000, blit=True)
```

### 3.1 - Urne d'Erhenfest

Soit $N \geq 1$, et $(Z^N)$ la chaine d'Erhenfest donné par le noyau de transition suivant: \\

$$
P_N(i,j) = \left\{
    \begin{array}{lll}
        \frac{i}{N} & \mbox{si } \ i = j+1 \\
        \frac{N-i}{N} & \mbox{si } \ i = j-1 \\
        0 & \mbox{sinon}
    \end{array}
\right.
$$


On construit alors la chaine $(Y_n^N)_{n\geq0} \ / \forall n \in \mathbb{N}, \ Y_n^N = \sqrt{N} (\frac{Z_n^N}{N} - \frac{1}{2})$. \\On peut donc maintenant appliquer le théorème d'approximation diffusive pour voir le comportement de la chaine de markov.


```python
fig = plt.figure(figsize=(11, 11))
ax = fig.gca()

max_T = 8
N = 20
PN = np.zeros((N+1,N+1))

T = np.linspace(0,max_T,100)

def XN(N,PN):
  #return 2*np.random.binomial(1,PN,N)-1
  return np.random.randn(N)
Positions = []
all_ = []

#ax.contour(X, Y, Z, alpha=.5)

def init():
    global Positions
    global N,PN,T
    global all_
    
    Positions = []
    all_ = []
    
    N = 2000
    PN = np.zeros((N+1,N+1))
    for i in range(N):
      PN[i+1,i] = float(i+1)/N
      PN[i,i+1] = float(N - i)/N
    return all_
    
def update(frame):
    global Positions
    global N,PN,T
    global all_
    #[a.remove() for a in all_]
    #all_.clear()
    
    #N *= 2
    ##PN = 0.5
    L = [int(0.8*N)]
    for i in range(int(max_T*N)):
      #print(L[-1],PN[:,L[-1]])
      L.append(np.random.choice(range(N+1), p = PN[L[-1],:]))
      
    pos = np.zeros(T.shape[0])
    for i in range(T.shape[0]):
      pos[i] = np.sqrt(N) * (L[int(T[i]*N)]/float(N) - 0.5)
      
    all_ += ax.plot(T,pos)
    
    return all_

animation.FuncAnimation(fig, update, init_func=init, frames=20, interval=1000, blit=True)
```
